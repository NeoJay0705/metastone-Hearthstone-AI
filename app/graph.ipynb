```python
for numberOfExperiment in range(1, 7):
    
    fitness_total = [0 for _ in range(8)]
    string_fileName = "experiment_0" + str(numberOfExperiment) + ".txt"
    with open(string_fileName) as file:
        # 必須設定
        int_populationSize = 5

        # 可有可無，方便觀察
        int_generation = 0
        float_fitness = float(0)

        # 必要變數，需重製置
        int_populationCount = 0
        list_X = []
        list_Y = []

        for data in file.readlines():
            data = data.replace("\n", "")
            data = data.strip()
            list_words = data.split(" ")

            for word in list_words:
                if word == "Generation":
                    int_generation = int(list_words[1])
                elif word == "F":
                    int_populationCount += 1
                    list_X.append(int_generation)
                    float_fitness = float(list_words[1])
                    list_Y.append(float_fitness)
                    fitness_total[int(float_fitness/10)] += 1


            if int_populationCount == int_populationSize:
                #print(list_X)
                #print(list_Y)
                plt.scatter(list_X, list_Y, s=10)
                plt.xlabel("Generation") 
                plt.ylabel("Fitness") 
                string_fileName = string_fileName.replace(".txt", "")
                plt.title(string_fileName) 
                int_populationCount = 0
                list_X = []
                list_Y = []
        plt.savefig(string_fileName,dpi=300,format="jpg")
        plt.show()
    print(fitness_total)
```
